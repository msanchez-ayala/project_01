```python
# Create a table for the movie information

cursor.execute("""
    CREATE TABLE tmdb (
    movie_id varchar(20) NOT NULL UNIQUE,
    imdb_id varchar(20) NOT NULL UNIQUE,
    title varchar(100) NOT NULL,
    budget int(12),
    revenue int(12),
    roi float(6),
    release_date varchar(10),
    PRIMARY KEY (movie_id)
    )
""")
```

```python
# Grab top 1000 movie ids (20 results per page). Increament over pages

def get_ids(genre):
    movies = []
    page = 1
    while page <= 50:
        url = 'https://api.themoviedb.org/3/discover/movie?api_key={}&with_genres={}&sort_by=revenue.desc&page={}'.format(api_key, genre, page)
        response = requests.get(url)
        moviez = response.json()['results']
        movies.extend(moviez)
        print(f"Grabbed page {page}")
        page += 1
        time.sleep(0.25)
        
    movie_ids = [movie['id'] for movie in movies]
    return movie_ids
```

```python
# get revenue and budget info for each movie in 'horror_movies'

def api_call():
    horror_movies = []
    count = 1
    for i_d in movie_ids:
        url = 'https://api.themoviedb.org/3/movie/{}?api_key={}&language=en-US&append_to_response=revenue'.format(i_d,api_key)
        response = requests.get(url)
        print(response.status_code)
        horror_moviez = response.json()
        horror_movies.append(horror_moviez)
        remaining_movies = 1000 - count
        print(f" {remaining_movies} movies remaining")
        count += 1
        time.sleep(0.25)
```

```python
# parse results for what we want

def parsed_results(movies):
    parsed_movies = []
    for movie in movies:
        if movie['imdb_id'] == None or movie['imdb_id'] == '':
            continue
        else:
            parsed = {}
            parsed['movie_id'] = movie['id']
            parsed['imdb_id'] = movie['imdb_id']
            parsed['title'] = movie['title']
            parsed['budget'] = movie['budget']
            parsed['revenue'] = movie['revenue']
            parsed['release_date'] = movie['release_date']
            parsed_movies.append(parsed)
    return parsed_movies
```

```python
parsed_horror_movies = parsed_results(horror_movies)
```

```python
def roi(movies):
    for movie in movies:
        if movie['budget'] != 0:
            movie['roi'] = round((movie['revenue'] / movie['budget'])-1, 2)
        else:
            movie['roi'] = 0.0
    return movies
```

```python
HorrorMovies = roi(parsed_horror_movies)
len(HorrorMovies)
```

```python
def del_budget(movies):
    parsed_movies = []
    for movie in movies:
        if movie['budget'] != 0:
            parsed_movies.append(movie)
    return parsed_movies
```

```python
def del_Nurse3D(movies):
    parsed_movies = []
    for movie in movies:
        if movie['title'] != 'Nurse 3-D':
            parsed_movies.append(movie)
    return parsed_movies
```
