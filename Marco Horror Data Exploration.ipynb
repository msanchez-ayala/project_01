# CREATE DATAFRAMES

```python
# MAKE IMDB/TMDB JOINED DF FOR HORROR

def create_tmdb_imdb(genre):
    """
    Returns a dataframe that joins the tmdb and imdb tables of any genre on imdb_id. Filters out low budget
        movies because wealthy investors probably are not interested in returns for budgets less than $50,000.
        Additionally filters out the Blair Witch Project (roi is above 400,000 %), which is an outlier.
    Param genre: [str] name of genre all lowercase.
    """
    cursor.execute(f"""
                   SELECT 
                     t.imdb_id,
                     t.title,
                     t.budget,
                     t.revenue,
                     t.roi AS 'ROI / %',
                     YEAR(t.release_date) AS release_year,
                     i.imdb_rank,
                     i.imdb_score,
                     i.metascore,
                     i.votes,
                     i.length
                   FROM 
                     tmdb_{genre} t
                   INNER JOIN
                     imdb_{genre} i
                   ON
                     t.imdb_id = i.imdb_id
                   WHERE
                     (t.budget > 50000)
                     AND
                     (t.roi < 400000)
                   ;""")
    result = pd.DataFrame(cursor.fetchall())
    result.columns = [x[0] for x in cursor.description]
    return result

def create_tmdb_imdb_outliers(genre):
    """
    Returns a dataframe that joins the tmdb and imdb tables of any genre on imdb_id. Filters out low budget
        movies because wealthy investors probably are not interested in returns for budgets less than $50,000.
    Param genre: [str] name of genre all lowercase.
    """
    cursor.execute(f"""
                   SELECT 
                     t.imdb_id,
                     t.title,
                     t.budget,
                     t.revenue,
                     t.roi AS 'ROI / %',
                     YEAR(t.release_date) AS release_year,
                     i.imdb_rank,
                     i.imdb_score,
                     i.metascore,
                     i.votes,
                     i.length
                   FROM 
                     tmdb_{genre} t
                   INNER JOIN
                     imdb_{genre} i
                   ON
                     t.imdb_id = i.imdb_id
                   WHERE
                     (t.budget > 50000)
                   ;""")
    result = pd.DataFrame(cursor.fetchall())
    result.columns = [x[0] for x in cursor.description]
    return result

def create_genre_other(table, genre):
    """
    Returns a DataFrame joined between genre and another table. Filters out low budget
        movies because wealthy investors probably are not interested in returns for budgets less than $50,000.
        Additionally filters out the Blair Witch Project (roi is above 400,000 %), which is an outlier.
    Param table: [str] Name of MySQL table to inner join on.
    Para genre: [str] Movie genre all lowercase.
    """
    cursor.execute(f"""
                   SELECT 
                     *
                   FROM 
                     genres_{genre} g
                   INNER JOIN
                     {table}_{genre} j
                   ON
                     g.imdb_id = j.imdb_id
                   WHERE
                     (j.budget > 50000)
                     AND
                     (j.roi < 400000)
                   ;""")
    result = pd.DataFrame(cursor.fetchall())
    result.columns = [x[0] for x in cursor.description]
    return result


def create_top_30(table, genre):
    """
    Returns a DataFrame of Top 30 by ROI. joined between tmdb and another table for a given genre.
        Filters out low budget movies because wealthy investors probably are not interested in 
        returns for budgets less than $50,000. Additionally filters out the Blair Witch Project 
        (roi is above 400,000 %), which is an outlier.
    Param table: [str] Name of MySQL table to inner join on.
    Para genre: [str] Movie genre all lowercase.
    """
    cursor.execute(f"""
                   SELECT
                     t.imdb_id,
                     t.title,
                     COUNT(*),
                     AVG(t.roi) AS avg_roi,
                     j.{table[:-1]}
                   FROM
                     movies.tmdb_{genre} t
                   INNER JOIN
                     movies.{table}_{genre} j
                   ON
                     t.imdb_id = j.imdb_id
                   WHERE
                     (t.budget > 50000)
                     AND
                     (t.roi < 400000)
                   GROUP BY
                     j.{table[:-1]}
                   ORDER BY
                     AVG(t.roi) DESC
                   LIMIT 30
                   ;
                   """)

    result = pd.DataFrame(cursor.fetchall())
    result.columns = [x[0] for x in cursor.description]
    return result
```

```python
tmdb_imdb_horror = create_tmdb_imdb('horror')
tmdb_imdb_horror_outliers = create_tmdb_imdb_outliers('horror')
```

```python
box_fig1 = plt.figure(figsize=(15,14))
box_ax1 = plt.subplot()

tmdb_imdb_horror_outliers['ROI / %'].plot(kind = 'box',
                          fontsize = 16,
                          ax=box_ax1
                          );
box_ax1.set_title('Return on Investment of Horror Movies',{'fontsize':16});
```

```python
box_fig2 = plt.figure(figsize=(15,10))
box_ax2 = plt.subplot()

tmdb_imdb_horror[tmdb_imdb_horror['ROI / %'] < 3000]['ROI / %'].plot(kind = 'box',
                          fontsize=16,
                          ax=box_ax2);
box_ax2.set_title('Return on Investment of Horror Movies',fontsize=16);
```
