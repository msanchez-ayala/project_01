# Horror Data Exploration

```python
import mysql.connector 
import marco
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
cnx = mysql.connector.connect(
    host = marco.host,
    user = marco.user,
    passwd = marco.password,
    database = 'movies'
)
cursor = cnx.cursor()
```

# CREATE DATAFRAMES

```python
# MAKE IMDB/TMDB JOINED DF FOR HORROR

def create_tmdb_imdb(genre):
    """
    Returns a dataframe that joins the tmdb and imdb tables of any genre on imdb_id. Filters out low budget
        movies because wealthy investors probably are not interested in returns for budgets less than $50,000.
        Additionally filters out the Blair Witch Project (roi is above 400,000 %), which is an outlier.
    Param genre: [str] name of genre all lowercase.
    """
    cursor.execute(f"""
                   SELECT 
                     t.imdb_id,
                     t.title,
                     t.budget,
                     t.revenue,
                     t.roi AS 'ROI / %',
                     YEAR(t.release_date) AS release_year,
                     i.imdb_rank,
                     i.imdb_score,
                     i.metascore,
                     i.votes,
                     i.length
                   FROM 
                     tmdb_{genre} t
                   INNER JOIN
                     imdb_{genre} i
                   ON
                     t.imdb_id = i.imdb_id
                   WHERE
                     (t.budget > 50000)
                     AND
                     (t.roi < 400000)
                   ;""")
    result = pd.DataFrame(cursor.fetchall())
    result.columns = [x[0] for x in cursor.description]
    return result

def create_tmdb_imdb_outliers(genre):
    """
    Returns a dataframe that joins the tmdb and imdb tables of any genre on imdb_id. Filters out low budget
        movies because wealthy investors probably are not interested in returns for budgets less than $50,000.
    Param genre: [str] name of genre all lowercase.
    """
    cursor.execute(f"""
                   SELECT 
                     t.imdb_id,
                     t.title,
                     t.budget,
                     t.revenue,
                     t.roi AS 'ROI / %',
                     YEAR(t.release_date) AS release_year,
                     i.imdb_rank,
                     i.imdb_score,
                     i.metascore,
                     i.votes,
                     i.length
                   FROM 
                     tmdb_{genre} t
                   INNER JOIN
                     imdb_{genre} i
                   ON
                     t.imdb_id = i.imdb_id
                   WHERE
                     (t.budget > 50000)
                   ;""")
    result = pd.DataFrame(cursor.fetchall())
    result.columns = [x[0] for x in cursor.description]
    return result

def create_genre_other(table, genre):
    """
    Returns a DataFrame joined between genre and another table. Filters out low budget
        movies because wealthy investors probably are not interested in returns for budgets less than $50,000.
        Additionally filters out the Blair Witch Project (roi is above 400,000 %), which is an outlier.
    Param table: [str] Name of MySQL table to inner join on.
    Para genre: [str] Movie genre all lowercase.
    """
    cursor.execute(f"""
                   SELECT 
                     *
                   FROM 
                     genres_{genre} g
                   INNER JOIN
                     {table}_{genre} j
                   ON
                     g.imdb_id = j.imdb_id
                   WHERE
                     (j.budget > 50000)
                     AND
                     (j.roi < 400000)
                   ;""")
    result = pd.DataFrame(cursor.fetchall())
    result.columns = [x[0] for x in cursor.description]
    return result


def create_top_30(table, genre):
    """
    Returns a DataFrame of Top 30 by ROI. joined between tmdb and another table for a given genre.
        Filters out low budget movies because wealthy investors probably are not interested in 
        returns for budgets less than $50,000. Additionally filters out the Blair Witch Project 
        (roi is above 400,000 %), which is an outlier.
    Param table: [str] Name of MySQL table to inner join on.
    Para genre: [str] Movie genre all lowercase.
    """
    cursor.execute(f"""
                   SELECT
                     t.imdb_id,
                     t.title,
                     COUNT(*),
                     AVG(t.roi) AS avg_roi,
                     j.{table[:-1]}
                   FROM
                     movies.tmdb_{genre} t
                   INNER JOIN
                     movies.{table}_{genre} j
                   ON
                     t.imdb_id = j.imdb_id
                   WHERE
                     (t.budget > 50000)
                     AND
                     (t.roi < 400000)
                   GROUP BY
                     j.{table[:-1]}
                   ORDER BY
                     AVG(t.roi) DESC
                   LIMIT 30
                   ;
                   """)

    result = pd.DataFrame(cursor.fetchall())
    result.columns = [x[0] for x in cursor.description]
    return result
```

```python
tmdb_imdb_horror = create_tmdb_imdb('horror')
tmdb_imdb_horror_outliers = create_tmdb_imdb_outliers('horror')
```

# Raw ROI Data

```python
box_fig1 = plt.figure(figsize=(15,14))
box_ax1 = plt.subplot()

tmdb_imdb_horror_outliers['ROI / %'].plot(kind = 'box',
                          fontsize = 16,
                          ax=box_ax1
                          );
box_ax1.set_title('Return on Investment of Horror Movies',{'fontsize':16});
```

## Zoomed-in, Filtered Horror Movies ROI Box Plot

```python
box_fig2 = plt.figure(figsize=(15,10))
box_ax2 = plt.subplot()

tmdb_imdb_horror[tmdb_imdb_horror['ROI / %'] < 3000]['ROI / %'].plot(kind = 'box',
                          fontsize=16,
                          ax=box_ax2);
box_ax2.set_title('Return on Investment of Horror Movies',fontsize=16);
```

```python
tmdb_imdb_horror['ROI / %'].median()
```

This box plot shows that we have a big spread of ROIs. Many outliers.

```python
# Kind of a useless graph. Doesn't give any new information that we didn't already have with the box plots


comparison_fig = plt.figure(figsize = (15,10),frameon=False);
comparison_ax1 = plt.subplot()


tmdb_imdb_horror[tmdb_imdb_horror['ROI / %']<3000].plot(x='length',
                      y='ROI / %',
                      kind='scatter',
                      ax=comparison_ax1,
                      color = 'teal',
                      fontsize=16)

comparison_ax1.set_xlim(70,175)
comparison_ax1.set_title('Return on Investment of Horror Movies vs. Movie Length',fontsize=16)
comparison_ax1.set_xlabel('Movie Length / minutes',fontsize=16), comparison_ax1.set_ylabel('ROI / %',fontsize=16);
```

Most data clustered between movie length 80 - 120 minutes. Seems we could commonly expect ROI below betwen -100 - 500.


# Subgenre Exploration

```python
tmdb_genres = create_genre_other('tmdb','horror')
```

```python
tmdb_genres.head()
```

```python
genres_grouped = tmdb_genres.groupby('genre').agg({'roi':['mean']}).drop('Horror')
```

```python
genres_fig = plt.figure(figsize = (18,12));
genres_ax1 = plt.subplot()

genres_grouped.plot(kind='bar',
                    ax=genres_ax1,
                    fontsize=16,
                    color = ['teal'],
                    legend=False)


genres_ax1.set_title('Average Return on Investment for Horror Movies by Subgenre',fontsize=16)
genres_ax1.set_xlabel('Genre',fontsize=16), genres_ax1.set_ylabel('ROI / %',fontsize=16);

```

# Stars Exploration

```python
top_30_stars = create_top_30('stars', 'horror').sort_values(by='avg_roi',ascending=False)
```

```python
top_30_stars.head()
```

```python
star_grid = (2,2) # Make a 2x2 axis grid to plot two graphs on top of each other in the same figure
star_fig = plt.figure(figsize = (22,14),frameon=False);
star_ax1 = plt.subplot2grid(star_grid, (0, 0),colspan=2) # First ax in positions (0,0) and (0,1) (top row)
star_ax2 = plt.subplot2grid(star_grid, (1, 0),colspan=2) # Second ax in positions (1,0) and (1,1) (bottom row)

top_30_stars.plot(x='star',
                  y='COUNT(*)',
                  kind='bar',
                  ax=star_ax1,
                  color = '#E37C2A',
                  fontsize=16,
                  legend=False);
star_ax1.set_title('No. of Appearances by Top 30 Horror Stars',fontsize=16);
star_ax1.set_xlabel('Name of Star',fontsize=16); 
star_ax1.set_ylabel('No. of Appearances',fontsize=16);
star_ax1.set_ylim(0,3);

top_30_stars.plot(x='star',
                  y='avg_roi',
                  kind='bar',
                  fontsize=16,
                  ax=star_ax2,
                  sharex=True,
                  legend=False,
                  color = 'teal',
                  rot=90);
star_ax2.set_title('Average Return on Investment of Horror Movies by Top 30 Horror Stars',fontsize=16);
star_ax2.set_xlabel('Name of Star',fontsize=16);
star_ax2.set_ylabel('ROI / %',fontsize=16);

```

Looks like you basically have some big outliers for the immensely popular films. Generally, all 'stars' appear only one time. This is likely because of the niche nature of the horror film market. >75% of the 2339 stars entries we have only appear once. This data is not reliable to make good recommendations.
